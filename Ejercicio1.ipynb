```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()
```

```python
conteo = data['pais'].value_counts().head(10)
```

```python
conteo
```

```python
# -----------------------
p = figure(x_range=list(conteo.index),
           title="Top 10 países con más aeropuertos",
           width=800, height=400)

p.vbar(x=list(conteo.index),
       top=conteo.values,
       width=0.6)

p.xaxis.major_label_orientation = 1
```

```python
show(p)
```
